```python
import pandas as pd

# Read the CSV file and select only the first column
DOC = pd.read_csv("processed_data/WoodsLakeMiddle_DissolvedOrganicCarbon_profile_Data.csv", encoding='unicode_escape')
DOC = DOC.loc[:,["Date","Data"]]
DOC["Date"] = DOC["Date"].apply(pd.to_datetime, format='%d/%m/%Y %H:%M', errors='coerce')
# Sort the DataFrame by 'Date' column
DOC = DOC.sort_values(by='Date')
print(DOC)

NPOC_dis = pd.read_csv("processed_data/WoodsLakeMiddle_DissolvedOrganicCarbonNon-Purgeable_profile_Data.csv", encoding='unicode_escape')
NPOC_dis = NPOC_dis.loc[:,["Date","Data"]]
NPOC_dis["Date"] = NPOC_dis["Date"].apply(pd.to_datetime, format='%d/%m/%Y %H:%M', errors='coerce')
NPOC_dis = NPOC_dis.sort_values(by='Date')
print(NPOC_dis)

NPOC_tot = pd.read_csv("processed_data/WoodsLakeMiddle_TotalOrganicCarbonNon-Purgeable_profile_Data.csv", encoding='unicode_escape')
NPOC_tot = NPOC_tot.loc[:,["Date","Data"]]
NPOC_tot["Date"] = NPOC_tot["Date"].apply(pd.to_datetime, format='%d/%m/%Y %H:%M', errors='coerce')
NPOC_tot = NPOC_tot.sort_values(by='Date')
print(NPOC_tot)
```

```python
import matplotlib.pyplot as plt

# Convert the 'Date' columns to datetime
DOC['Date'] = pd.to_datetime(DOC['Date'])
NPOC_dis['Date'] = pd.to_datetime(NPOC_dis['Date'])
NPOC_tot['Date'] = pd.to_datetime(NPOC_tot['Date'])

# Filter the dataframes to include only the data from 2015 to 2024
start_date = '2000-01-01'
end_date = '2024-12-31'

DOC_filtered = DOC[(DOC['Date'] >= start_date) & (DOC['Date'] <= end_date)]
NPOC_dis_filtered = NPOC_dis[(NPOC_dis['Date'] >= start_date) & (NPOC_dis['Date'] <= end_date)]
NPOC_tot_filtered = NPOC_tot[(NPOC_dis['Date'] >= start_date) & (NPOC_tot['Date'] <= end_date)]

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(DOC_filtered['Date'], DOC_filtered['Data'], label='DOC')
plt.plot(NPOC_dis_filtered['Date'], NPOC_dis_filtered['Data'], label='NPOC_dis')
plt.plot(NPOC_tot_filtered['Date'], NPOC_tot_filtered['Data'], label='NPOC_tot')

plt.xlabel('Date')
plt.ylabel('OC')
plt.legend()
plt.grid(True)
plt.show()
```
