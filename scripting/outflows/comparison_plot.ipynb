```python
import pandas as pd

# Read the CSV file and select only the first column
spillway = pd.read_csv("Interpolated_462.1_WoodsLakeAtDam_SpillwayDischarge_Continuous.csv", encoding='unicode_escape')
print(spillway)
lakeriver = pd.read_csv("Interpolated_454.1_LakeRiverBLWoodsLakeDam_ChannelFlow_Continuous.csv", encoding='unicode_escape')
print(lakeriver)
```

```python
import matplotlib.pyplot as plt

# Convert the 'Date' columns to datetime
spillway['Date'] = pd.to_datetime(spillway['Date'])
lakeriver['Date'] = pd.to_datetime(lakeriver['Date'])

# Filter the dataframes to include only the data from 2015 to 2024
start_date = '2015-01-01'
end_date = '2024-12-31'

spillway_filtered = spillway[(spillway['Date'] >= start_date) & (spillway['Date'] <= end_date)]
lakeriver_filtered = lakeriver[(lakeriver['Date'] >= start_date) & (lakeriver['Date'] <= end_date)]

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(spillway_filtered['Date'], spillway_filtered['Data'], label='Spillway Discharge')
plt.plot(lakeriver_filtered['Date'], lakeriver_filtered['Data'], label='Lake River Channel Flow')

plt.xlabel('Date')
plt.ylabel('Discharge/Flow')
plt.title('Comparison of Spillway Discharge and Lake River Channel Flow (2015-2024)')
plt.legend()
plt.grid(True)
plt.show()
```
