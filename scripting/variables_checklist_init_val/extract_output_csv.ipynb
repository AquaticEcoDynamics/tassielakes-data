```python
# import OS module
import os
# Get the list of all files and directories
paths = ["../../data-warehouse/csv/ht/alwq",
        "../../data-warehouse/csv/ht/met",
        "../../data-warehouse/csv/ht/wlwq"
        ]
dir_lists = []
for path in paths:
    dir_list = os.listdir(path)
    print(len(dir_list))
    dir_lists += dir_list
    print("Files and directories in '", path, "' :")
    # prints all files
    print(dir_list)

print(dir_lists)
print(len(dir_lists))
```

```python
filename = []
for dir in dir_lists:
    filename.append(dir.split('_')[0] + '_' + dir.split('_')[1])

filename.remove('.DS_Store')
print(filename)
print(len(filename))
```

```python
file = list(dict.fromkeys(filename))
file.remove('.DS_Store')
print(file)
print(len(file))
```

```python
source = []
variable = []
for i in file:
    source.append(i.split('_')[0])
    variable.append(i.split('_')[1])

import pandas as pd
df = pd.DataFrame(list(zip(source, variable)), columns=['Source', 'Variable'])
df["File"] = '.csv'
print(df)
```

```python
df.to_csv("variable_in_csv_outputs.csv", index = False)
```
