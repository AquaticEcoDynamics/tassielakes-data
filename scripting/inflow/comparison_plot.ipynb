```python
import pandas as pd

# Read the CSV file
spillway = pd.read_csv("Interpolated_418.1_ArthursSpillway.csv", encoding='unicode_escape')
print(spillway)
inflow = pd.read_csv("Interpolated_418.1_ArthursInflow.csv", encoding='unicode_escape')
print(inflow)
```

```python
import matplotlib.pyplot as plt

# Convert the 'Date' columns to datetime
spillway['Date'] = pd.to_datetime(spillway['Date'])
inflow['Date'] = pd.to_datetime(inflow['Date'])

# Filter the dataframes to include only the data from 2015 to 2024
start_date = '2015-01-01'
end_date = '2024-12-31'

spillway_filtered = spillway[(spillway['Date'] >= start_date) & (spillway['Date'] <= end_date)]
inflow_filtered = inflow[(inflow['Date'] >= start_date) & (inflow['Date'] <= end_date)]

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(spillway_filtered['Date'], spillway_filtered['Data'], label='Spillway Discharge')
plt.plot(inflow_filtered['Date'], inflow_filtered['Data'], label='Lake River Channel Inflow')

plt.xlabel('Date')
plt.ylabel('Discharge/Flow')
plt.title('Comparison of Spillway Discharge and Lake River Channel Inflow (2015-2024)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Read the CSV file
inflow_lakeriver = pd.read_csv("Interpolated_418.1_ArthursInflow.csv", encoding='unicode_escape')
print(inflow_lakeriver)

inflow_212_1_140 = pd.read_csv("Interpolated_212-1-140_Inflow.csv", encoding='unicode_escape')
print(inflow_212_1_140)

inflow_SWIFT_natural = pd.read_csv("Interpolated_SWIFTnatural_Inflow.csv", encoding='unicode_escape')
print(inflow_SWIFT_natural)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert the 'Date' columns to datetime
inflow_lakeriver['Date'] = pd.to_datetime(inflow_lakeriver['Date'])
inflow_212_1_140['Date'] = pd.to_datetime(inflow_212_1_140['Date'])
inflow_SWIFT_natural['Date'] = pd.to_datetime(inflow_SWIFT_natural['Date'])

# Filter the dataframes to include only the data from 2015 to 2024
start_date = '2015-01-01'
end_date = '2024-12-31'

inflow_lakeriver_filtered = inflow_lakeriver[(inflow_lakeriver['Date'] >= start_date) & (inflow_lakeriver['Date'] <= end_date)]
inflow_212_1_140_filtered = inflow_212_1_140[(inflow_212_1_140['Date'] >= start_date) & (inflow_212_1_140['Date'] <= end_date)]
inflow_SWIFT_natural_filtered = inflow_SWIFT_natural[(inflow_SWIFT_natural['Date'] >= start_date) & (inflow_SWIFT_natural['Date'] <= end_date)]

# Plot the data with more contrasting colors
plt.figure(figsize=(14, 7))
plt.plot(inflow_lakeriver_filtered['Date'], inflow_lakeriver_filtered['Data'], label='inflow_lakeriver', color='blue')
plt.plot(inflow_SWIFT_natural_filtered['Date'], inflow_SWIFT_natural_filtered['Data'], label='inflow_SWIFT_natural', color='green')
plt.plot(inflow_212_1_140_filtered['Date'], inflow_212_1_140_filtered['Data'], label='inflow_212_1_140', color='red')

plt.xlabel('Date')
plt.ylabel('Discharge/Flow')
plt.title('Comparison (2015-2024)')
plt.legend()
plt.grid(True)
plt.show()
```
