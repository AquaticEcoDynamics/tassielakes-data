```python
import pandas as pd

# Read the CSV file
spillway = pd.read_csv("Interpolated_418.1_ArthursSpillway.csv", encoding='unicode_escape')
print(spillway)
inflow = pd.read_csv("Interpolated_418.1_ArthursInflow.csv", encoding='unicode_escape')
print(inflow)
```

```python
import matplotlib.pyplot as plt

# Convert the 'Date' columns to datetime
spillway['Date'] = pd.to_datetime(spillway['Date'])
inflow['Date'] = pd.to_datetime(inflow['Date'])

# Filter the dataframes to include only the data from 2015 to 2024
start_date = '2015-01-01'
end_date = '2024-12-31'

spillway_filtered = spillway[(spillway['Date'] >= start_date) & (spillway['Date'] <= end_date)]
inflow_filtered = inflow[(inflow['Date'] >= start_date) & (inflow['Date'] <= end_date)]

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(spillway_filtered['Date'], spillway_filtered['Data'], label='Spillway Discharge')
plt.plot(inflow_filtered['Date'], inflow_filtered['Data'], label='Lake River Channel Inflow')

plt.xlabel('Date')
plt.ylabel('Discharge/Flow')
plt.title('Comparison of Spillway Discharge and Lake River Channel Inflow (2015-2024)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files
arthursinflow = pd.read_csv("flowIntoLakeRiverArthurs_inflow_woods_dam_20160107-20240514.csv", usecols=[0,1], encoding='unicode_escape')
outflow = pd.read_csv("../outflow/outflow_woods_dam-spillway_19990311-20240416.csv", encoding='unicode_escape')
swiftinflow = pd.read_csv("swift_inflow_woods_dam_20160107-20240514.csv", usecols=[0,1], encoding='unicode_escape')
storagenetinflow = pd.read_csv("Interpolated_storageNet_Inflow.csv", encoding='unicode_escape')
wl = pd.read_csv("../../data-warehouse/csv/ht/wlwq/WoodsLakeMiddleContinuous_WaterSurfaceHeight_profile_Data.csv")

# Rename columns for consistency
arthursinflow.columns = ['Date', 'Flow']
outflow.columns = ['Date', 'Flow']
swiftinflow.columns = ['Date', 'Flow']
storagenetinflow.columns = ['Date', 'Flow']

# Convert the 'Date' columns to datetime with the correct format
arthursinflow['Date'] = pd.to_datetime(arthursinflow['Date'], format='%d/%m/%Y %H:%M:%S')
outflow['Date'] = pd.to_datetime(outflow['Date'], format='%d/%m/%Y %H:%M:%S')
swiftinflow['Date'] = pd.to_datetime(swiftinflow['Date'], format='%d/%m/%Y %H:%M:%S')
storagenetinflow['Date'] = pd.to_datetime(storagenetinflow['Date'], format='%Y-%m-%d %H:%M:%S')
wl['Date'] = pd.to_datetime(wl['Date'], format='%Y-%m-%d %H:%M:%S')

# Filter the dataframes to include only the data from 2015 to 2024
start_date = '2023-01-01'
end_date = '2024-07-01'

arthursinflow_filtered = arthursinflow[(arthursinflow['Date'] >= start_date) & (arthursinflow['Date'] <= end_date)]
outflow_filtered = outflow[(outflow['Date'] >= start_date) & (outflow['Date'] <= end_date)]
outflow_filtered['Flow'] *= -1
swiftinflow_filtered = swiftinflow[(swiftinflow['Date'] >= start_date) & (swiftinflow['Date'] <= end_date)]
storagenetinflow_filtered = storagenetinflow[(storagenetinflow['Date'] >= start_date) & (storagenetinflow['Date'] <= end_date)]
wl_filtered = wl[(wl['Date'] >= start_date) & (wl['Date'] <= end_date)]
print(wl_filtered)
# Plot the data with more contrasting colors
fig, ax = plt.subplots(figsize=(14, 7))
ax2 = ax.twinx()
ax.plot(storagenetinflow_filtered['Date'], storagenetinflow_filtered['Flow'], label='Storage Net Inflow', color='lightgrey')
ax.plot(swiftinflow_filtered['Date'], swiftinflow_filtered['Flow'], label='SWIFT inflow', color='green')
ax.plot(arthursinflow_filtered['Date'], arthursinflow_filtered['Flow'], label='Arthurs Inflow', color='blue')
ax.plot(outflow_filtered['Date'], outflow_filtered['Flow'], label='Outflow', color='red')

ax2.plot(wl_filtered['Date'],wl_filtered['Data'],label='Water Level', color='black')

# Combine legends for both axes
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc='upper right')

ax.set_ylim(-4, 30)

ax.set_xlabel('Date')
ax.set_ylabel('Discharge/Flow (cumecs)')
ax2.set_ylabel('Water Level (m)')
plt.title('Comparison (2015-2024)')
# plt.grid(True)
plt.show()
```
