```python
import pandas as pd

phycocyanin = pd.read_csv("../../../data-warehouse/csv/ht/wlwq/WoodsLakeMiddleSurfaceContinuous_Phycocyanin_profile_Data.csv")
print(phycocyanin)

cyanobacteria = pd.read_csv("../../../data-warehouse/csv/ht/wlwq/WoodsLakeMiddle_PhytoplanktonBiomass(diag_cyano)_profile_Data.csv")
print(cyanobacteria)
```

```python
import matplotlib.pyplot as plt

phycocyanin['Date'] = pd.to_datetime(phycocyanin['Date'], format='%Y-%m-%d %H:%M:%S')
cyanobacteria['Date'] = pd.to_datetime(cyanobacteria['Date'], format='%Y-%m-%d %H:%M:%S')

start_date = '2023-01-01'
end_date = '2024-01-01'

phycocyanin_filtered = phycocyanin[(phycocyanin['Date'] >= start_date) & (phycocyanin['Date'] <= end_date)]
print(phycocyanin_filtered)
cyanobacteria_filtered = cyanobacteria[(cyanobacteria['Date'] >= start_date) & (cyanobacteria['Date'] <= end_date)]
print(cyanobacteria_filtered)

# Plot the data
fig, ax = plt.subplots(figsize=(14,7))
ax2 = ax.twinx()

ax.plot(phycocyanin_filtered['Date'], phycocyanin_filtered['Data'], label='phycocyanin', color='black')
ax2.plot(cyanobacteria_filtered['Date'], cyanobacteria_filtered['Data'], label='cyanobacteria', color='red')

# Combine legends for both axes
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc='upper right')

ax.set_xlabel('Date')
ax.set_ylabel('Phycocyanin (pigment)')
ax2.set_ylabel('Cyanobacteria (mmol/m^3)')

plt.title('Correlation Comparison')

plt.show()
```
